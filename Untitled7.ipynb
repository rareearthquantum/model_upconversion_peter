```python
#importing python libs

import sympy as sym
sym.init_printing()

import numpy as np
from math import pi
import math
import matplotlib.pyplot as plt
from sympy import I, Matrix, symbols
from sympy.physics.quantum import TensorProduct, Dagger
import scipy.optimize
import scipy.integrate
import scipy.constants as const

#import qutip

from matplotlib.colors import Normalize as Norm

%load_ext cython
```

```python
#define some s pre/post operators

def spre(m):
    return TensorProduct(sym.eye(m.shape[0]),m)

def spost(m):
    return TensorProduct(m.T, sym.eye(m.shape[0]))

def collapse(c):
    tmp = Dagger(c)*c/2
    return spre(c)*spost(Dagger(c))-spre(tmp)-spost(tmp)


s13=Matrix([[0,0,1],[0,0,0],[0,0,0]])
s23=Matrix([[0,0,0],[0,0,1],[0,0,0]])
s12=Matrix([[0,1,0],[0,0,0],[0,0,0]])

s31=s13.T
s32=s23.T
s21=s12.T

s11 = s12*s21
s22 = s21*s12
s33 = s31*s13
```

```python
deltao,deltamu=sym.symbols('delta_o delta_mu', real=True) #detunings for the levels 2 and 3
gamma13,gamma23,gamma2d,gamma3d,nbath,gammamu=sym.symbols('gamma_13 gamma_23 gamma_2d gamma_3d n_b gamma_mu', real=True, negative=False) #energy decay for atom levels
Omega=sym.symbols('Omega', real=True, negative=False) #pump Rabi frequency
rho11, rho12, rho13, rho21, rho22, rho23, rho31, rho32, rho33=sym.symbols('rho_11 rho_12 rho_13 rho_21 rho_22 rho_23 rho_31 rho_32 rho_33') #Density matrix elements
Delo, Delmu= sym.symbols('Delta_o Delta_mu',real=True)
a, b = sym.symbols('a b') #classical amplitudes of the optical and microwave fields
#ar,ai=sym.symbols('a_r a_i', real=True)
go, gmu=sym.symbols('g_o, g_mu',real=True, negative=False) #coupling strengths for optical and microwave fields
lam=sym.symbols('lambda')
```

```python
H_sys=Omega*s32+gmu*s21*b+go*s31*a
H_sys=H_sys+Dagger(H_sys)
H_sys=H_sys+(deltamu+Delmu)*s33+(deltao+Delo)*s22
```

```python
LH=-I*spre(H_sys)+I*spost(H_sys)
L21 = gammamu*(nbath+1)*collapse(s12)
L12 = gammamu*nbath*collapse(s21)
L32 = gamma23*collapse(s23)
L31 = gamma13*collapse(s13)
L22 = gamma2d*collapse(s22)
L33 = gamma3d*collapse(s33)

L=LH + L21 + L12 + L32 + L31 + L22 + L33

L = L.row_insert(0,Matrix([[1,0,0,0,1,0,0,0,1]]))
L.row_del(1)

#define the density matrix in square and row form
#the row form is so the Liovillian in matrix form can be acted on it
rho = Matrix([[rho11,rho21,rho31],[rho12,rho22,rho32],[rho13,rho23,rho33]])
rho = 1*rho.T #because we are using "fortran" style matrix flatteneing
rho[:]
rhoflat = 1*rho.T 
rhoflat = rhoflat[:]
```

```python
Lfunc = sym.lambdify((a,b,Delo, Delmu,deltao, deltamu, gamma13, gamma23, gamma2d, gamma3d, nbath,gammamu,Omega,go,gmu),L)
```

```python
def steady_rho_single(delta2val,delta3val,aval, bval,Deloval,Delmuval,p):

    L_mat=Lfunc(aval, bval,Deloval,Delmuval, delta2val,delta3val,p['gamma13'],p['gamma23'],p['gamma2d'],p['gamma3d'], p['nbath'],p['gammamu'],p['Omega'],p['go'],p['gmu'])
    rho = np.linalg.solve(L_mat.astype('complex128'),np.matrix([[1,0,0,0,0,0,0,0,0]]).T)
    return np.reshape(rho,(3,3),order='F')
def steady_rho_ensemble(delta2vals, delta3vals,aval,bval,Deloval,Delmuval, p):
    ndelta2=len(delta2vals)
    ndelta3=len(delta3vals)
    coh_vals=np.zeros((3,3,ndelta2,ndelta3), dtype=np.complex_)
    for ii in range(ndelta2):
        for jj in range(ndelta3):
            coh_vals[:,:,ii,jj]=steady_rho_single(delta2vals[ii],delta3vals[jj],aval,bval,Deloval,Delmuval,p)
    return coh_vals
```

```python
def integrate1d_peaks(func,x,ylims, peak_func,aval,bval,Deloval,Delmuval,p):
    peak_yval=peak_func(x)
    
    if (np.isnan(peak_yval) or peak_yval>ylims[1] or peak_yval<ylims[0]):
        funcval=scipy.integrate.quad(func, ylims[0], ylims[1], args=(x,aval,bval,Deloval,Delmuval,p))[0]
        
    else:
        funcval1=scipy.integrate.quad(func, ylims[0], peak_yval, args=(x,aval,bval,Deloval,Delmuval,p))[0]
        funcval2=scipy.integrate.quad(func, peak_yval, ylims[1], args=(x,aval,bval,Deloval,Delmuval,p))[0]
        funcval=funcval1+funcval2
    return funcval

def integrate2d_peaks(func, ylims, xlims, peak_func,aval,bval,Deloval,Delmuval,p):
    temp_fun=lambda x: integrate1d_peaks(func, x, ylims, peak_func,aval,bval,Deloval,Delmuval,p)
    inte=scipy.integrate.quad(temp_fun,xlims[0], xlims[1])
    return inte
```

```python
coh13_Rfun=lambda del2, del3, aval1=aval, bval1=bval,Deloval1=Deloval,Delmuval1=Delmuval,p=p : np.real(steady_rho_single(del2,del3,aval1,bval1,Deloval1,Delmuval1,p)[0,2])
coh13_Ifun=lambda del2, del3, aval1=aval, bval1=bval,Deloval1=Deloval,Delmuval1=Delmuval,p=p : np.imag(steady_rho_single(del2,del3,aval1,bval1,Deloval1,Delmuval1,p)[0,2])
gauss_fun=lambda del2, del3, m2,m3,sd2,sd3: 1.0/(2*pi*sd2*sd3)*np.exp(-(del2-m2)**2/(2*sd2**2)
                                                                  -(del3-m3)**2/(2*sd3**2))
coh13_Rguass_fun=lambda del2,del3,aval1=aval,bval1=bval,Deloval1=Deloval,Delmuval1=Delmuval,p=p,m2=p['mean_delta2'],m3=p['mean_delta3'],sd2=p['sd_delta2'],sd3=p['sd_delta3']: coh13_Rfun(del2,del3,aval1,bval1,Deloval1,Delmuval1,p)*gauss_fun(del2,del3,m2,m3,sd2,sd3)
coh13_Iguass_fun=lambda del2,del3,aval1=aval,bval1=bval,Deloval1=Deloval,Delmuval1=Delmuval,p=p,m2=p['mean_delta2'],m3=p['mean_delta3'],sd2=p['sd_delta2'],sd3=p['sd_delta3']: coh13_Ifun(del2,del3,aval1,bval1,Deloval1,Delmuval1,p)*gauss_fun(del2,del3,m2,m3,sd2,sd3)
```

```python
coh12_Rfun=lambda del2, del3, aval1=aval, bval1=bval,Deloval1=Deloval,Delmuval1=Delmuval,p=p : np.real(steady_rho_single(del2,del3,aval1,bval1,Deloval1,Delmuval1,p)[0,1])
coh12_Ifun=lambda del2, del3, aval1=aval, bval1=bval,Deloval1=Deloval,Delmuval1=Delmuval,p=p : np.imag(steady_rho_single(del2,del3,aval1,bval1,Deloval1,Delmuval1,p)[0,1])
coh12_Rguass_fun=lambda del2,del3,aval1=aval,bval1=bval,Deloval1=Deloval,Delmuval1=Delmuval,p=p,m2=p['mean_delta2'],m3=p['mean_delta3'],sd2=p['sd_delta2'],sd3=p['sd_delta3']: coh12_Rfun(del2,del3,aval1,bval1,Deloval1,Delmuval1,p)*gauss_fun(del2,del3,m2,m3,sd2,sd3)
coh12_Iguass_fun=lambda del2,del3,aval1=aval,bval1=bval,Deloval1=Deloval,Delmuval1=Delmuval,p=p,m2=p['mean_delta2'],m3=p['mean_delta3'],sd2=p['sd_delta2'],sd3=p['sd_delta3']: coh12_Ifun(del2,del3,aval1,bval1,Deloval1,Delmuval1,p)*gauss_fun(del2,del3,m2,m3,sd2,sd3)
```

```python
def rho13_integrated(del2_lims, del3_lims,aval,bval,Deloval,Delmuval,p, peak_func):
    coh13_R_int=integrate2d_peaks(coh13_Rguass_fun, del2_lims,del3_lims,peak_func,aval,bval,Deloval,Delmuval,p)
    coh13_I_int=integrate2d_peaks(coh13_Iguass_fun, del2_lims,del3_lims,peak_func,aval,bval,Deloval,Delmuval,p)
    #return [coh13_R_int, coh13_I_int]#coh13_R_int+1.j*coh13_I_int\
    #return coh13_R_int+1.j*coh13_I_int
    return coh13_R_int[0]+1.j*coh13_I_int[0],coh13_R_int[1]+1.j*coh13_I_int[1]
def rho13_integrated_quad(del2_lims, del3_lims,aval,bval,Deloval,Delmuval,p):
    coh13_R_int=scipy.integrate.nquad(coh13_Rguass_fun, [del2_lims,del3_lims],args=(aval,bval,Deloval,Delmuval,p))[0]
    coh13_I_int=scipy.integrate.nquad(coh13_Iguass_fun, [del2_lims,del3_lims],args=(aval,bval,Deloval,Delmuval,p))[0]
    #return [coh13_R_int, coh13_I_int]#coh13_R_int+1.j*coh13_I_int\
    return coh13_R_int+1.j*coh13_I_int
```

```python
def rho12_integrated(del2_lims, del3_lims,aval,bval,Deloval,Delmuval,p, peak_func):
    coh12_R_int=integrate2d_peaks(coh12_Rguass_fun, del2_lims,del3_lims,peak_func,aval,bval,Deloval,Delmuval,p)
    coh12_I_int=integrate2d_peaks(coh12_Iguass_fun, del2_lims,del3_lims,peak_func,aval,bval,Deloval,Delmuval,p)
    #return [coh13_R_int, coh13_I_int]#coh13_R_int+1.j*coh13_I_int\
    #return coh13_R_int+1.j*coh13_I_int
    return coh12_R_int[0]+1.j*coh12_I_int[0],coh12_R_int[1]+1.j*coh12_I_int[1]
def rho12_integrated_quad(del2_lims, del3_lims,aval,bval,Deloval,Delmuval,p):
    coh12_R_int=scipy.integrate.nquad(coh12_Rguass_fun, [del2_lims,del3_lims],args=(aval,bval,Deloval,Delmuval,p))[0]
    coh12_I_int=scipy.integrate.nquad(coh12_Iguass_fun, [del2_lims,del3_lims],args=(aval,bval,Deloval,Delmuval,p))[0]
    #return [coh13_R_int, coh13_I_int]#coh13_R_int+1.j*coh13_I_int\
    return coh12_R_int+1.j*coh12_I_int
```

```python
# calculate optical rabi frequency from power in in dBm
def omega_from_Pin(Pin,p):
    epsilon0=8.854187817e-12
    hbar=1.05457e-34; # in J*s

    optP = 1e-3 * 10**(Pin/10) #incident optical power in W
    pflux = optP/(2*pi*p['freq_pump']*hbar) #photon flux (photons/sec)
    n_in = pflux * p['kappaoc']*4/(p['kappaoc']+p['kappaoi'])**2 # num intracavity photons
                                                #is this right????
    Sspot = pi*p['Wcavity']**2 #cross sectional area of
                                #optical mode
    V_cav = (Sspot*p['Lcavity_vac']+Sspot*p['Lsample']*p['nYSO']**3)/2;
    optEfield = math.sqrt(n_in*hbar*2*pi*p['freq_pump']/2/epsilon0/V_cav);
    p['Omega'] = p['d23']*optEfield/hbar*(-1);
    return p['Omega']
```

```python
def a_from_Pin(Pin,p):

    epsilon0=8.854187817e-12
    hbar=1.05457e-34; # in J*s
    optP=1e-3*10**(Pin/10)
    pflux=optP/(2*pi*p['freqo']*hbar)
    n_in = pflux*p['kappaoc']*4/(p['kappaoc']+p['kappaoi'])**2
    Sspot = pi*p['Wcavity']**2
    V_cav= (Sspot*p['Lcavity_vac']+Sspot*p['Lsample']*p['nYSO']**3)/2 # what
    optEfield=np.sqrt(n_in*hbar*2*pi*p['freqo']*2/epsilon0/V_cav)
    omegao=p['d13']*optEfield/hbar*(-1)
    a_return = omegao/p['go']
    return a_return
```

```python
def a_from_Pin2(Pin,p):
    epsilon0=8.854187817e-12
    hbar=1.05457e-34;
    Sspot = pi*p['Wcavity']**2 #cross sectional area of
                                #optical mode
    V_cav = (Sspot*p['Lcavity_vac']+Sspot*p['Lsample']*p['nYSO']**3)/2;
    optP=1e-3*10**(Pin/10)
    Q=2*pi*p['freqo']/(p['kappaoi']+p['kappaoc']) # Q factor make sure this is correct, check with wikipedia
    S21=(4*p['kappaoc']**2)/(p['kappaoi']+p['kappaoc']*2)**2
    e_in_cavity=np.sqrt(S21)*2/(2*pi*p['freqo']/Q)*optP #make sure correct
    Eopt=np.sqrt(2/epsilon0*e_in_cavity/V_cav)
    omegao=p['d13']*Eopt/hbar*(-1)
    a_return=omegao/p['go']
    return a_return
    
```

```python
def b_from_Pin(Pin,p):
    mu0 = 4*pi*1e-7
    hbar=1.05457e-34; # in J*s
    Vsample=pi*((p['dsample']/2)**2) * p['Lsample'] # the volume of the Er:YSO sample
    V_microwave_cavity = Vsample/p['fillfactor']    
    mwP = 1e-3 * 10**(Pin/10)  #convert from dBm to Watts
    Q=2*pi*p['freqmu']/(p['kappami']+p['kappamc']*2) # Q factor
    S21=(4*p['kappamc']**2)/(p['kappami']+p['kappamc']*2)**2
    energy_in_cavity=math.sqrt(S21)*2/(2*pi*p['freqmu']/Q)*mwP; # energy inside the microwave cavity, in J
    Bmw=math.sqrt(mu0*(energy_in_cavity/V_microwave_cavity)/2); # Magnetic field of the microwave
    omegam=(p['mu12']*Bmw)/hbar*(-1); # in rad/sec???? in Hz
    b_return=omegam/p['gmu']
    return b_return
```
